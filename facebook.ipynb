```python
%matplotlib inline
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from random import randint
import pip
```

```python
import networkx as nx
G = nx.read_edgelist("/content/drive/MyDrive/facebook_1000_data.txt")

node = input("Enter the node: ")
all_nodes = set(G.nodes)
neighbors_set = set(G.neighbors(node))
non_neighbors = all_nodes - neighbors_set - {node}

def common_neighbors(u, v):
  return len(list(nx.common_neighbors(G, u, v)))

def jaccard_coefficient(u, v):
  scores = list(nx.jaccard_coefficient(G, [(u, v)]))
  return scores[0][2]

def preferential_attachment(u, v):
  scores = list(nx.preferential_attachment(G, [(u, v)]))
  return scores[0][2]

print("Select a similarity score:")
print("1. Common Neighbors")
print("2. Jaccard's Coefficient")
print("3. Preferential Attachment")

choice = int(input())

user = node

max_score = -float('inf')
friend_suggestion = None

for neighbor in non_neighbors:
  if choice == 1:
    score = common_neighbors(node, neighbor)  
  elif choice == 2:
    score = jaccard_coefficient(node, neighbor)
  elif choice == 3:
    score = preferential_attachment(node, neighbor)

  if score >= max_score:
    max_score = score
    friend_suggestion = neighbor


  if score >= max_score:
    max_score = score
    friend_suggestion = neighbor


print(f"Friend suggestion for {user}: {friend_suggestion}")
```
